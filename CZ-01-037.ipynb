```python
import matplotlib.pyplot as plt
import pandas as pd
df1 = pd.read_csv("D:/UTAustin/DSC/CZ-01-037.csv")
x1=df1['Temperature']
y1=df1['Heat Flow']
```

```python
Afont = {'fontname':'Arial'}
plt.figure(figsize = (5,4))
plt.plot(x1, y1, label = 'PCEGE_3.9 kg/mol ', color = 'darkorange', linewidth=2)
plt.xlim(-82,2)
plt.ylim(-0.47,-0.08)
plt.xlabel(u'Temperature (\u00B0C)', **Afont, fontsize='14')
plt.ylabel('Heat Flow (W/g)', **Afont, fontsize='14')
plt.xticks(**Afont, fontsize='12')
plt.yticks(**Afont, fontsize='12')
ax=plt.gca()
ax.tick_params(left=False, labelleft=False)
ax.yaxis.labelpad = 10
plt.legend(loc=(0.35,0.85), prop = {'family':'Arial', 'size':'14'}, frameon=False)
plt.vlines(-37.625, -0.32, -0.26, color='black', linestyle = '-', linewidth=2)
plt.text(-50, -0.42, r'$T_g$' + ' = \N{MINUS SIGN}37.62' + u' (\u00B0C)', **Afont, fontsize='14')
plt.text(-80, -0.19, r'$\rightarrow$'+' Heating (10\u00B0C/min)', **Afont, fontsize='14')
plt.text(-95, -0.46, 'Exo Up', **Afont, fontsize='12')
plt.savefig('CZ-01-037_DSC.tiff', dpi = 200)
plt.show()
```
