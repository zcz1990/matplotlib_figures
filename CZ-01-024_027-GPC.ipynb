```python
import matplotlib.pyplot as plt
import pandas as pd
df1 = pd.read_csv("D:/UTAustin/GPC/CZ-01-027_PAGE.csv")
x1=df1['RT (mins)']
y1=df1['Response Trace 4']
df2 = pd.read_csv("D:/UTAustin/GPC/CZ-01-024_PAGE.csv")
x2=df2['RT (mins)']
y2=df2['Response Trace 4']
fig, ax1 = plt.subplots()
```

```python
Afont = {'fontname':'Arial'}
plt.figure(figsize = (5,4))
plt.plot(x1, y1, label = 'PAGE_9.4 k', color = 'darkslateblue', linewidth=2)
plt.xlim(12.5,17.5)
plt.ylim(-100,40000)
plt.xlabel('Retention Time (mins)', **Afont, fontsize='14')
plt.ylabel('RI Intensity', **Afont, fontsize='14')
plt.xticks(**Afont, fontsize='12')
plt.yticks(**Afont, fontsize='12')
plt.legend(loc=(0.0,0.7), prop = {'family':'Arial', 'size':'14'}, frameon=False)
ax1 = plt.gca()
ax1.axes.yaxis.set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)
ax1.spines['left'].set_visible(False)

ax2 = ax1.twinx()

plt.plot(x2, y2, label = 'PAGE_17.8 k', color = 'darkorange', linewidth=2)
plt.xlim(12.5,17.5)
plt.ylim(-100,48000)
plt.xlabel('Retention Time (mins)', **Afont, fontsize='14')
plt.ylabel('RI Intensity', **Afont, fontsize='14')
plt.xticks(**Afont, fontsize='12')
plt.yticks(**Afont, fontsize='12')
plt.legend(loc=(0.0,0.6), prop = {'family':'Arial', 'size':'14'}, frameon=False)

ax2 = plt.gca()
ax2.axes.yaxis.set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax2.spines['left'].set_visible(False)
plt.savefig('CZ-01-024_027-GPC.tiff', dpi = 300)
plt.show()
```
